```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
movies_data = pd.read_csv(r'C:\Users\huitd\Downloads\movies.csv')
movies_data.head()
```

```python
movies_data.shape
```

```python
# выбор соответствующих столбцов для рекомендательных систем
selected_features = ['genres', 'keywords', 'tagline', 'cast', 'director']
```

```python
# замена нулевых значений нулевой строкой
for feature in selected_features:
    movies_data[feature] = movies_data[feature].fillna('')
```

```python
# обьединим все выбранные столбцы

combined_features = movies_data['genres']+' '+movies_data['keywords']+' '+movies_data['tagline']+' '+movies_data['cast']+' '+movies_data['director']               
```

```python
print(combined_features)
```

```python
# преобразование текстовых данных в векторы признаков

vectorizer = TfidfVectorizer()
```

```python
feature_vectors =vectorizer.fit_transform(combined_features)
```

```python
print(feature_vectors)
```

```python
# получение оценок сходства с помощью cosin_simmilarity

similarity = cosine_similarity(feature_vectors)
similarity
```

```python
similarity.shape
```

```python
# получение названия фильма от пользователя

movie_name = input('Enter the movie name: ')
```

```python
# создание списка со всеми фильмами, указанными в наборе данных

list_of_all_titles = movies_data['title'].tolist()
print(list_of_all_titles)
```

```python
# находим близкое соответствие фильму, указанному пользователем

find_the_close_match = difflib.get_close_matches(movie_name, list_of_all_titles)
find_the_close_match
```

```python
close_match = find_the_close_match[0]
close_match
```

```python
# находим индекс фильма с названием

index_of_the_movie = movies_data[movies_data.title == close_match]['index'].values[0]
index_of_the_movie
```

```python
# получаем список похожих фильмов

similarity_score = list(enumerate(similarity[index_of_the_movie]))
similarity_score
```

```python
len(similarity_score)
```

```python
# сортировка фильмов на основе их оценки сходства

sorted_simillar_movies = sorted(similarity_score, key = lambda x:x[1], reverse=True)
sorted_simillar_movies
```

```python
# выводим название похожих фильмов

print('Movies suggested to you: ')

i = 1

for movie in sorted_simillar_movies:
    index = movie[0]
    title_from_index = movies_data[movies_data.index==index]['title'].values[0]
    if (i<30):
        print(i, '-', title_from_index)
        i+=1
```
